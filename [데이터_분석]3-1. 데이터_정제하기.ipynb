## 데이터 정제 Data Cleaning
데이터 처리(Data Wrangling) 또는 데이터 변형(Data Munging)은 데이터를 수집하고 저장된 형식에서 분석에 적합한 형식으로 변환하고 정리하는 과정을 의미합니다. 이는 데이터 과학 및 데이터 분석 작업에서 매우 중요한 단계 중 하나입니다.

### 데이터 처리(Data Wrangling)의 주요 단계:

1. **데이터 수집**: 데이터를 수집하여 분석할 수 있는 형식으로 가져옵니다. 이는 웹 스크래핑, API 호출, 데이터베이스 쿼리 등 다양한 방법으로 수행될 수 있습니다.

2. **데이터 정제**: 데이터에 결측치, 이상치, 중복된 값 등이 있는 경우 이를 처리하고 데이터를 정리합니다. 이를 통해 데이터의 품질을 향상시키고 분석 결과의 신뢰도를 높일 수 있습니다.

3. **데이터 변환**: 데이터를 분석에 적합한 형식으로 변환합니다. 예를 들어, 날짜 형식의 데이터를 파싱하여 날짜 및 시간 객체로 변환하거나, 범주형 데이터를 원-핫 인코딩 등의 방법으로 수치형 데이터로 변환합니다.

4. **데이터 통합**: 여러 소스에서 가져온 데이터를 하나의 데이터 집합으로 통합하고 병합합니다. 이는 데이터베이스 조인과 유사한 개념으로, 특정 기준에 따라 데이터를 결합합니다.

5. **데이터 형식 변환**: 데이터를 다양한 형식으로 변환하거나 저장합니다. 예를 들어, CSV 파일을 데이터베이스에 저장하거나, 데이터베이스에서 가져온 데이터를 JSON 형식으로 변환합니다.

### 데이터 처리(Data Wrangling)의 중요성:
- 실제 데이터는 종종 불완전하고 지저분합니다. 데이터 처리를 통해 이러한 불완전성과 불일치성을 해결하여 신뢰할 수 있는 분석 결과를 얻을 수 있습니다.
- **데이터 처리는 분석 및 모델링 과정에서 가장 많은 시간을 소요하는 작업 중 하나입니다. 따라서 효율적인 데이터 처리는 전체 프로젝트의 성공에 중요한 역할을 합니다.**
- 데이터 처리는 데이터 과학 및 기계 학습 모델의 성능을 향상시키는 데 필수적입니다. 잘 처리된 데이터는 모델의 정확도를 향상시키고 예측 성능을 개선할 수 있습니다.

요약하면, 데이터 처리 또는 데이터 변형은 데이터를 분석에 적합한 형식으로 변환하고 정리하는 과정으로, 데이터 과학 및 데이터 분석 프로젝트에서 매우 중요한 단계 중 하나입니다.

NumPy 배열(numpy array)과 파이썬 리스트(Python list)는 데이터를 저장하고 조작하는 데 사용되는 두 가지 주요 데이터 구조입니다. 이들 간에는 몇 가지 중요한 차이가 있습니다.

1. **메모리 사용 및 성능:**
   - NumPy 배열은 고정된 데이터 타입을 갖기 때문에 파이썬 리스트보다 메모리 사용이 효율적입니다. 또한 NumPy 배열은 C로 구현되어 있어 연산이 빠르고 효율적입니다.
   - 파이썬 리스트는 다양한 데이터 타입을 포함할 수 있고, 각 항목이 포인터로 연결되어 있기 때문에 메모리 사용이 더 많고 연산이 더 느릴 수 있습니다.

2. **데이터 타입:**
   - NumPy 배열은 동일한 데이터 타입의 요소만을 포함할 수 있습니다. 이는 데이터를 보다 효율적으로 저장하고 연산을 수행하는 데 도움이 됩니다.
   - 파이썬 리스트는 서로 다른 데이터 타입의 요소를 포함할 수 있습니다.

3. **연산 및 브로드캐스팅:**
   - NumPy 배열은 배열 간의 연산과 브로드캐스팅을 지원합니다. 따라서 같은 크기의 배열 간에 간단한 산술 연산을 수행할 수 있습니다.
   - 파이썬 리스트는 산술 연산이나 브로드캐스팅을 지원하지 않으며, 요소별로 연산을 수행하기 위해 반복문이 필요합니다.

4. **인덱싱과 슬라이싱:**
   - NumPy 배열은 다차원 배열로 구성될 수 있으며, 다차원 인덱싱 및 슬라이싱을 지원합니다.
   - 파이썬 리스트는 단일 차원으로만 구성되며, 슬라이싱을 통해 부분 리스트를 선택할 수 있지만 다차원 인덱싱은 지원하지 않습니다.

따라서 NumPy 배열은 수치 계산과 데이터 분석과 같은 과학 및 엔지니어링 응용 프로그램에서 효율적으로 사용되며, 대량의 데이터를 처리하거나 복잡한 연산을 수행해야 할 때 유용합니다. 반면에 파이썬 리스트는 일반적인 프로그래밍에서 다양한 데이터 타입을 다루는 데 유용하며, 리스트를 조작하는 데 유연성이 필요한 경우에 사용됩니다.

# 그룹별로 모으기

````python
count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
````

````python
# 기본적으로 groupby는 by 매개변수에 지정된 열에 nan 값이 포함되어 있으면 해당 행을 삭제
# 따라서 nan이 있어도 값을 포함시켜 계산하려면 dropna = False로 설정
group_df = count_df.groupby(by = ['도서명', '저자', 'ISBN', '권'], dropna = False)
loan_count = group_df.sum()
````

> ### update() 메서드는 두 개의 DataFrame을 사용하여 특정 조건을 만족하는 값들을 한 DataFrame에서 다른 DataFrame으로 업데이트합니다. 이때, 두 DataFrame의 인덱스는 동일하거나 같은 라벨을 공유해야 합니다. 하지만 행의 순서가 다르더라도 상관없습니다.

- update() 메서드는 인덱스가 동일한 위치에 있는 값들을 서로 매칭하여 업데이트합니다. 만약 첫 번째 DataFrame에 해당 인덱스가 없는 경우, 해당 인덱스는 그대로 유지되고 아무런 변화가 일어나지 않습니다.

```python
import pandas as pd
# 첫 번째 DataFrame
df1 = pd.DataFrame({'A': [1, 2, 3]}, index=['a', 'b', 'c'])
# 두 번째 DataFrame
df2 = pd.DataFrame({'A': [4, 5]}, index=['b', 'c'])
# update() 메서드를 사용하여 업데이트
df1.update(df2)
print(df1)
```
- 위 코드에서 첫 번째 DataFrame df1의 인덱스 'b'와 'c'에 해당하는 값들이 두 번째 DataFrame df2에서의 값으로 업데이트됩니다. 'a'에 해당하는 값은 df2에 없기 때문에 변화가 없습니다. 따라서 update() 메서드는 두 DataFrame을 업데이트할 때, 인덱스가 동일한 위치에 있는 값들을 서로 매칭하여 업데이트합니다.

> **정리**
- 그냥 기존 dataframe에는 다른 column들도 있기에, 대출 건수를 합해줄 기준을 만들어줄 column들만 뽑아서, 계산을 하고, 기존 원본 데이터 프레임에 인덱스를 아까 뽑은 column들과 맞추어서, column의 값들을 update하고 다시 인덱스를 없애주었다.

````python

def data_cleaning(filename):
    """
    남산 도서관 장서 CSV 데이터 전처리 함수

    :param filename: CSV 파일이름
    """
    # 파일을 데이터프레임으로 읽습니다.
    ns_df = pd.read_csv(filename, low_memory=False)
    # NaN인 열을 삭제합니다.
    ns_book = ns_df.dropna(axis=1, how='all')

    # 대출건수를 합치기 위해 필요한 행만 추출하여 count_df 데이터프레임을 만듭니다.
    count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
    # 도서명, 저자, ISBN, 권을 기준으로 대출건수를 groupby합니다.
    loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
    # 원본 데이터프레임에서 중복된 행을 제외하고 고유한 행만 추출하여 복사합니다.
    dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])
    unique_rows = ~dup_rows
    ns_book3 = ns_book[unique_rows].copy()
    # 도서명, 저자, ISBN, 권을 인덱스로 설정합니다.
    ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)
    # load_count에 저장된 누적 대출건수를 업데이트합니다.
    ns_book3.update(loan_count)

    # 인덱스를 재설정합니다.
    ns_book4 = ns_book3.reset_index()
    # 원본 데이터프레임의 열 순서로 변경합니다.
    ns_book4 = ns_book4[ns_book.columns]

    return ns_book4
````
