- **p-value**
- p-value에서의 "p"는 "probability"의 약자입니다. p-value는 통계적 가설 검정에서 사용되는 중요한 개념 중 하나입니다. 주어진 데이터에 대해 귀무 가설을 검정할 때, p-value는 귀무 가설이 참이라는 전제 아래에서 관측된 데이터 또는 더 극단적인 결과가 나타날 확률을 나타냅니다. 즉, p-value가 작을수록 우리가 귀무 가설을 기각하고 대립 가설을 채택할 확신을 갖게 됩니다.

- 일반적으로, p-value가 0.05보다 작으면 통계적으로 유의하다고 판단됩니다. 즉, 해당 결과를 얻을 확률이 5%보다 작다는 것을 의미합니다. 하지만 이 값은 특정한 상황과 분야에 따라 달라질 수 있습니다.

- 예를 들어, 특정 실험에서 p-value가 0.03이라면, 이는 해당 결과가 우연히 발생할 확률이 3% 미만이라는 것을 의미합니다. 따라서 이러한 결과는 통계적으로 유의하다고 할 수 있습니다.

- 반대로, p-value가 0.05보다 크다면, 우리는 귀무 가설을 기각하기 어려울 것이며, 우리가 얻은 결과가 우연에 의한 것일 수 있다는 가능성이 높아집니다.

- 통계학에서, 모집단에 대한 parameter(평균, 분산 등)를 추정하는 방법을 모수검정이라고 한다.
- 모집단 : 관심 대상이 되는 전체 데이터
- 전체 집단을 구하기란 어렵고, 불가능함
- 따라서, 일부 집단(sample 표본)을 조사한 다음 전체 집단에 대한 추정을 하게됨. 이것이 바로 "모수 검정"이다.

> 이때, 모수검정은 모집단의 데이터에 어떤 가정을 전제로 하고 수행하는 경우가 많음
- ex) 청소년의 몸무게가 정규분포를 따른다.
- 이러한 가정은 실제로 청소년의 몸무게가 정규분포를 따른다는 것이 사실인지는 모르지만, 자연 세계의 데이터는 정규분포를 따르는 경우가 많기 때문에 진행 <**[중심극한정리](https://blog.deeplink.kr/?p=2487)**>


> **표준 점수 구하기**
- 데이터가 정규분포를 따른다고 가정했을 때, 각 값들이 평균에서 얼마나 떨어져 있는지 표준편차를 사용해 변환한 점수를 **표준점수 or z 점수**라고 한다.
- ex) 출생한 아기의 몸무게와 키에 대한 데이터가 있다고 가정 -> 몸무게와 키가 평균으로부터 얼마만큼 떨어져 있는지 나타내보자.
- 몸무게는 1kg, 키는 5cm와 같이 서로 다른 단위로 나타내어야 한다. 게다가, 이런 수치만 봐서는 평균으로부터 몸무게가 더 멀리 떨어져 있는지, 키가 더 멀리 떨어져 있는 지 알 수 없다.
- 그래서 z 점수를 사용한다.

- if 아기의 몸무게와 키에 대한 z score가 1과 0.5라면, 아기의 몸무게가 평균에서 1표준편차만큼 떨어져 있고, 키는 0.5만큼 떨어져 있다는 의미

```python
# z score 구하기
import numpy as np

x = [0, 3, 5, 7, 10]

s = np.std(x)
m = np.mean(x)
z = (7 - m) / s
print(z)

# 7에 대한 Z score
```

```python
from scipy import stats
stats.zscore(x)
```

> **누적 분포**
- 평균이 0, 표준편차가 1인 정규 분포 : 표준 정규 분포
- 평균 0 , 표준편차 1을 z 점수 공식에 대입하면, z = x가 됨 => 표준 정규 분포는 z 점수에 따라 전체 데이터가 어떻게 분포되어 있는 지 나타낼 수 있음

- ex) 표준정규분포에서 z 점수가 1이내에 위치한 샘플은 전체의 68%에 해당한다.

```python
# 어떤 z 점수 이내의 샘플 비율을 계산하는 사이파이의 함수
# stats 모듈의 norm.cdf()
from scipy import stats

stats.norm.cdf(0)
# 누적 분포 : cumulated distribution function
```

## 중심 극한 정리 Central Limit Theorem
### 충분히 큰 표본의 표본평균은 근사적으로 정규 분포를 따른다. 이는 표본이 원래 어떤 분포를 따르든 상관없이 적용된다.

- 무작위로 샘플을 뽑아 만든 표본의 평균은 정규분포에 가깝다.

- 중심극한정리는 중요한 통계학적 개념 중 하나로, 표본 평균이 정규분포에 가까워지는 경향을 설명합니다. 자연계의 많은 데이터가 정규분포를 따른다는 주장은 중심극한정리의 일종으로 볼 수 있습니다. 이는 자연계에서 발생하는 다양한 현상들이 많은 개별 요인들의 합으로 나타나기 때문에, 중심극한정리에 따라 이들의 평균은 정규분포를 따르게 되는 것입니다.

- 따라서, 모수 검정을 수행할 때 표본 크기가 충분히 크다면 중심극한정리에 의해 표본 평균은 정규분포를 따를 가능성이 높습니다. 이는 모집단이 정규분포를 따르지 않더라도 표본 평균이 정규분포를 따를 수 있다는 것을 의미합니다. 이는 모수 검정에서 가정을 만족시키는 데 도움이 될 수 있습니다.

- 그러나 주의할 점은, 중심극한정리가 정확하게 적용되려면 표본 크기가 충분히 커야 합니다. 따라서 모수 검정을 수행할 때에도 적절한 표본 크기를 선택하여 중심극한정리가 적용될 수 있도록 해야 합니다.

> **이는 모집단의 분포에 대한 정보가 부족하거나 알 수 없더라도 표본 평균의 분포가 정규분포를 따른다는 것을 의미, 모집단이 어떤 분포를 따르는지에 대한 사전 정보가 없거나 정확하지 않은 경우에도 통계적 추론을 할 수 있게 해준다.** 다만, 극단적으로 비대칭적인 분포나 특이값이 있는 경우에는 주의가 필요하다.

### 표본의 평균으로부터 모집단의 평균 범위 추정하기 : 신뢰 구간

- 이전에는 30개의 data로 이루어진 1000개의 sample을 뽑아 계산한 표본의 평균이 모집단의 평균에 가까워진다는 것을 확인했다.
- 그러나, 현실에서는 표본을 수집하는 데 시간과 비용이 많이 들기에, 이렇게 많은 횟수의 표본을 만들기 어렵다
- 게다가 모집단에 대한 정보를 구할 수도 없다.

> if 딱 하나의 표본이 있다면 모집단의 평균은 어떻게 추정할까? => **신뢰구간** confidence interval

- 신뢰구간은 표본의 파라미터가 있을 것이라 생각되는 모집단의 파라미터 범위이다.

```python
python_books_index = ns_book7['주제분류번호'].str.startswith('00') & ns_book7['도서명'].str.contains('파이썬')
python_books = ns_book7[python_books_index]
python_books.head()
```

이때, z 점수 공식과 중심극한 정리를 연결해서 모집단의 평균 범위를 예측

1. 모집단의 평균을 알 수 없고, 표본의 평균만 아니까
2. z = x - 뮤 / s
3. z = (14.749 - 뮤) / (sigma / 루트 n)
- 남산 도서관의 파이썬 대출건수가 sample(표본)이라면 우리나라의 모든 도서관의 파이썬 도서 대출건수를 모집단으로 생각 가능
- 이때 중심 극한 정리를 사용 가능
4. 모집단의 표준편차가 표본의 표준편차와 비슷하다고 가정
5. 모집단의 표준편차 대신 표본이라 할 수 있는 남산 도서관의 파이썬 도서 대출건수로 표준편차를 구한 다음, 표준 오차를 계산
6. 뮤 = 14.719 - z *(sigma / 루트 n)

## 통계적 의미 확인하기 : 가설 검정

- 표본에 대한 정보를 사용해 모집단의 파라미터에 대한 가정을 검정하는 것
- ex) 파이썬과 C++ 도서의 평균 대출건수가 같다고 가정했을 때, 파이썬과 c++ 도서의 표본을 각각 추출하여 이 가정이 맞는지 검정 가능
- "파이썬과 c++ 도서의 평균 대출 건수가 같다는 가설" : 영가설 or 귀무 가설 null hypothesis
- "파이썬과 c++ 도서의 평균 대출건수가 같지 않다는 가설" : 대립 가설 alternative hypothesis


>1. **영가설 (Null Hypothesis)**: H0
   - 영가설은 일반적으로 연구자가 반박하거나 기각하고자 하는 가설입니다.
   - 주로 통계적 검정을 통해 기존에 받아들여진 가설이나 기준을 나타냅니다.
   - 예를 들어, "두 그룹의 평균이 같다"라는 주장은 영가설로 설정될 수 있습니다.
   - **표본 사이에 통계적으로 의미가 없다고 예상되는 가설**
2. **대립가설 (Alternative Hypothesis)**: HA
   - 대립가설은 영가설과 반대되는 주장이나 가설입니다.
   - 연구자가 검증하고자 하는 새로운 가설이나 예상을 나타냅니다.
   - 예를 들어, "두 그룹의 평균이 다르다"라는 주장은 대립가설로 설정될 수 있습니다.
   - **표본 사이에 통계적인 차이가 있다는 가설**

- 통계적 가설 검정에서는 주로 영가설을 기각하거나 받아들이는 것을 목표로 합니다. 즉, 표본 데이터를 사용하여 영가설에 대한 적절한 결론을 내리는 것이 주요 목적입니다. 이를 통해 대립가설이 영가설을 기각하거나 받아들이는지를 결정합니다.
- 가설 검정은 일반적으로 유의수준과 귀무가설을 설정하여 수행됩니다. 유의수준은 영가설을 기각하는 기준이 되는 임계값으로, 일반적으로 0.05 또는 0.01과 같은 값으로 설정됩니다. 유의수준을 기준으로 샘플 데이터의 분석 결과가 영가설을 기각할지 수용할지를 결정합니다.

- 영가설의 경우, 뮤1 - 뮤2가 0이라고 가정하고 z score를 계산
- **이때, 계산해서 나온 z score가 일정 수준 이하이면 0 가설이 맞다고 하고 임계값 이상이면, 이는 두 집단 간의 평균 차이가 없는 데이터가 관측될 확률이 95%보다 낮아, 귀무가설을 믿어 줄 수 없음을 의미한다. => 영가설을 기각하고 대립 가설을 채택함**

- 두 집단 간의 차이가 없다는 귀무가설을 기각하기 위해서는 주로 통계적인 유의성 수준을 설정하게 됩니다. 통상적으로 95%의 신뢰수준을 사용하는 경우가 많으며, 이는 5%의 유의수준을 가지게 됩니다. 이때 이 유의수준이라는 건, 5% 까지는 우연히 발생한 차이라고 봐줄 수 있으니 원래 주장하던 영가설을 채택해줄 수 있는 허용 범위이다. 따라서, 만약 두 집단 간의 차이가 통계적으로 유의미하다고 판단하려면, 해당 차이가 정규분포의 양 끝부분의 2.5% 구간에 해당하는 지 보면 된다. 이는 일반적으로 임계값을 기준으로 하여, 해당 차이의 z 점수가 임계값을 초과하면 귀무가설을 기각하고 대립가설을 채택하게 됩니다. 이는 95% 신뢰도에서 믿어줄 수 있는 구간을 넘은 차이를 보여, 두 집단 사이에 통계적으로 유의미한 차이가 있다는 것을 의미하며, 해당 결과가 우연히 발생한 것이 아닌 실제로 존재하는 차이임을 나타냅니다.
- 정규분포의 가장자리의 각 2.5% 까지의 차이는 두 집단이 서로 평균이 같은데도 우연히 발생한 차이라고 이해해 줄 수 있는데, 그 선을 넘어가면 그건 우연히 발생한 차이보다는 두 집단 사이에 차이가 있는 것이라고 보는게 맞아.

- 유의수준 : z 점수에 대한 기준으로, 일반적으로 정규분포의 양쪽 꼬리 면적을 더해 5%가 되는 지점

### t-검정으로 가설 검증하기
- scipy의 ttest_ind()는 t-분포인 두 표본을 비교하는 t-검정을 수행함
- t-분포는 정규분포와 비슷하지만, 중앙은 조금 더 낮고 꼬리가 두꺼운 분포
- 표본의 크기가 30 이하일 떄, t-분포를 사용하는 것이 좋고, 표본 크기가 30 이상이면 t-분포는 정규분포와 매우 비슷해짐
- 따라서, 이 함수를 사용해 표본 크기에 상관없이 평균 비교가 가능

t-test와 z-test는 두 집단 간의 평균 차이를 비교하는 데 사용되는 통계적 검정 방법입니다. 그러나 두 검정 방법에는 몇 가지 중요한 차이점이 있습니다.

1. **표본 크기와 모집단 표준편차의 여부**:
   - z-test는 모집단의 표준편차를 알고 있고, 표본의 크기가 큰 경우에 사용됩니다.
   - 반면에 t-test는 모집단의 표준편차를 모르고, 표본의 크기가 작은 경우에 사용됩니다. 특히 표본 크기가 30 미만인 경우에는 t-test를 사용하는 것이 일반적입니다.

2. **통계량의 분포**:
   - z-test는 표준 정규 분포를 따르는 z 통계량을 사용합니다.
   - t-test는 t 분포를 따르는 t 통계량을 사용합니다. t 분포는 표본 크기에 따라 모양이 변하는데, 표본 크기가 커질수록 표준 정규 분포에 가까워집니다.

3. **가정의 여부**:
   - z-test는 모집단의 분포가 정규 분포를 따른다는 가정을 전제로 합니다.
   - t-test는 모집단의 분포가 정규 분포를 따르지 않더라도 표본 크기가 충분하면 중심극한정리에 따라 t 분포를 따르는 것으로 간주됩니다.

따라서, z-test는 모집단의 표준편차를 알고 있고, 표본의 크기가 큰 경우에 사용되며, t-test는 모집단의 표준편차를 모르고, 표본의 크기가 작은 경우에 사용됩니다.

### **정규분포가 아닐 때의 가설 검증하기 : 순열 검정 permutation test**

- 모집단의 분포가 정규분포를 따르지 않거나 모집단의 분포를 알 수 없을 때 사용할 수 있는 방법
- 모집단의 parameter를 추정하지 않기에, 비모수검정 nonparametic test 방법 중 하나이다.

> 1. 먼저 두 표본의 평균의 차이를 계산한다.
2. 두 표본을 섞고, 무작위로 두 그룹으로 나눕니다.
3. 이떄 두 표본은 원래 표본의 size와 동일하게 만든다.
4. 이렇게 나눈 두 그룹에서 다시 평균의 차이를 계산
5. 이 과정을 여러번 반복해서 원래 표본의 평균 차이가 무작위로 나눈 그룹의 평균 차이보다 크거나 작은 경우를 헤아려 p-값을 계산한다.

순열 검정(Permutation Test)은 비모수적인 통계적 방법 중 하나로, 표본 데이터의 순서를 임의로 섞어서 새로운 표본을 생성하여 가설을 검정하는 방법입니다. 이 방법은 표본의 분포를 가정하지 않고, 관측된 데이터에 기초하여 유의수준을 추정하는 데 사용됩니다.

순열 검정은 다음과 같은 과정으로 진행됩니다:

1. **가설 설정**: 연구자는 검정하고자 하는 가설을 설정합니다. 일반적으로는 두 집단 간의 평균이 같다는 귀무가설과 이에 반하는 대립가설을 설정합니다.

2. **검정 통계량 계산**: 원래의 데이터에서 계산된 검정 통계량을 구합니다. 이는 주로 두 집단 간의 평균 차이를 나타내는 통계량입니다.

3. **표본의 순서 섞기**: 원래의 데이터에서 두 집단을 임의로 섞어 새로운 표본을 생성합니다. 이를 통해 귀무가설이 참일 때, 두 집단 간의 차이가 우연히 발생할 확률을 추정할 수 있습니다.

4. **검정 통계량 계산**: 새로운 순열로 생성된 표본에서 검정 통계량을 다시 계산합니다.

5. **퍼뮤테이션 분포 생성**: 순열을 여러 번 반복하여 검정 통계량의 분포를 생성합니다. 이를 통해 귀무가설이 참일 때 검정 통계량의 분포를 추정할 수 있습니다.

6. **p-value 계산**: 실제로 관측된 검정 통계량이 퍼뮤테이션 분포에서 어느 위치에 있는지를 확인하여 p-value를 계산합니다.

7. **결과 해석**: 계산된 p-value를 유의수준과 비교하여 귀무가설을 기각하거나 채택합니다.

순열 검정은 특히 모집단의 분포가 정규 분포를 따르지 않을 때, 또는 표본의 크기가 작은 경우에 유용하게 사용될 수 있습니다.

[순열 검정의 배경 이론과 절차](https://angeloyeo.github.io/2021/10/28/permutation_test.html)

[bootstrap과 순열 검정](https://syj9700.tistory.com/57)

```python
def statistic(x, y):
    return np.mean(x) - np.mean(y)
```

```python
# 두 배열을 numpy append()로 합친 후, 무작위로 추출하기 위해 permutation()함수를 사용
# permutation() : 무작위로 섞인 배열 생성
# 이번 함수에서는 전체 배열 길이만큼 랜덤한 인덱스를 생성

def permutation_test(x,y):
  # 표본의 평균 차이를 계산합니다.
  obs_diff = statistic(x,y)
  # 두 표본을 합칩니다.
  all = np.append(x,y)
  diffs = []
  np.random.seed(42)
  # 순열 검정을 1000번 반복합니다.
  for _ in range(1000):
    # 전체 인덱스를 섞는다.
    idx = np.random.permutation(len(all))
    # 랜덤하게 두 그룹으로 나눈 다음 평균 차이를 계산합니다
    x_ = all[idx[:len(x)]]
    y_ = all[idx[len(x):]]
    diffs.append(statistic(x_, y_))
  # 원본 표본보다 작거나 큰 경우의 p-값을 계산합니다.
  less_pvalue = np.sum(diffs < obs_diff)/1000
  greater_pvalue = np.sum(diffs > obs_diff)/1000
  # 둘 중 작은 p-값을 선택해 2를 곱하여 최종 p-값을 반환합니다.
  return obs_diff, np.minimum(less_pvalue, greater_pvalue) * 2
```

```python
permutation_test(python_books['대출건수'], cplus_books['대출건수'])
```

> **p-value : 영가설이 옳다고 가정했을 때 이러한 데이터가 관측될 확률**

 p-value는 주어진 데이터에서 관측된 결과가 귀무 가설이 참일 때 현재 가진 sample과 같은 데이터가 다시 발생할 수 있는 확률을 의미합니다. 따라서 p-value가 작을수록 귀무 가설이 맞다는 가정 아래에서 현재의 샘플과 유사한 데이터를 관측할 확률이 낮다는 것을 의미합니다.

예를 들어, p-value가 0.05라면, 해당 결과를 얻을 확률이 5%보다 낮다는 것을 의미합니다. 따라서 귀무 가설이 맞다는 전제 아래에서는 현재의 샘플과 유사한 데이터를 관측할 확률이 매우 낮다고 할 수 있습니다.

이것은 통계적으로 우리가 얻은 결과가 우연에 의한 것이 아니라는 것을 시사합니다. 그러므로 p-value가 작을수록 우리는 귀무 가설을 기각하고 대립 가설을 채택할 수 있는 확신을 가지게 됩니다.

### scipy의 permutation_test()

```python
# scipy 1.8 버전 이상에서만 실행됩니다.
res = stats.permutation_test((python_books['대출건수'], cplus_books['대출건수']),
                             statistic, random_state=42)
# 결과는 약 3.153 0.0258입니다.
print(res.statistic, res.pvalue)
```
