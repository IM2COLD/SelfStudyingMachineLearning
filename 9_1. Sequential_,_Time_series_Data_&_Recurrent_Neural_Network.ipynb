## **주요 개념 정리**

> 복잡한 model을 배울수록 가중치 개수를 계산해 보고, model.summary()를 통해 나온 가중치의 값과 비교해보며 model을 잘 이해하고 있는 지 알 수 있다!

#### 순차 data : 텍스트나 시계열 데이터와 같이 **순서에 의미가 있는 데이터**를 말한다.
- 이전의 데이터들과는 다르게 텍스트 data는 단어의 순서가 중요한 데이터이다.
- 이전에 입력한 data를 기억하는 기능이 필요하다
- ex) 별로지만 추천해요 => 추천해요만을 보고 긍정적이라고 판단하지 않도록, 별로지만을 기억해야함

#### **피드포워드 신경망 feedforward neural network**
- 입력 data의 흐름이 앞으로만 전달되는 신경망 => 완전 연결 신경망, 합성곱 신경망

- 이전에 처리했던 sample을 다음 sample을 처리하는데 재사용하기 위해 이전 데이터가 신경망 층에 순환될 필요가 있다. => **순환 신경망**

#### **순환 신경망**
- 완전 연결 신경망에 이전 data의 처리 흐름을 순환하는 고리 하나만 추가하면 된다.

> A, B, C 순서로 data가 들어올 때 -> A가 뉴런을 통과하여 Oa라는 출력이 나올 때, 출력값 Oa가 다시 뉴런으로 들어가서 이후에 B가 뉴런을 통과해서 Ob라는 출력이 나올 때 A에 대한 정보가 어느 정도 포함되어 있게 된다. 마찬가지로 C를 처리할 때에는 Ob가 계산에 활용되어 b와 a의 정보를 함께 사용하여 Oc를 출력하게 된다.

- Oc에는 b에 대한 정보가 a에 대한 정보보다 많이 담겨 있을 것이다.
- 순환신경망 : 이전 sample에 대한 기억을 가지고 있다.
- => 이렇게 **sample을 처리하는 하나의 단계 : TimeStep 타임스텝**이라고 말한다
- timestep이 오래될수록 순환되는 정보는 희미해진다.

**in 순환신경망 -> 층을 cell 셀이라고 부른다.**
- 한 셀에는 여러 개의 뉴런이 있지만 완전 연결 신경망과 달리 뉴런을 모두 표시하지 ㅇ낳고, 하나의 셀로 층을 표현
- 셀의 출력을 은닉 상태 hidden state라고 부른다.

> **순환 신경망은 피드포워드 신경망과 달리, 입력에 곱해지는 가중치와 이전 타임 스텝의 은닉상태에 곱해지는 가중치가 별도로 존재한다.**

- 타임 steo에서 은닉상태와 곱해지는 가중치느 모든 타임 스텝에서 동일하다.
- 순환층에 있는 하나의 뉴런의 은닉상태가 다음 time step에서 재사용될 때, 다른 뉴런들에게도 모두 전달된다. => **이전 타임 스텝의 은닉상태는 다음 timestep의 뉴런에 완전히 연결됨**

- 은닉 상태가 모든 뉴런에 순환되기 때문에, 완전 연결 신경망처럼 그림으로 표현하기는 어렵다. => 순환층 전체를 cell 하나로 표시함

> **순환층은 일반적으로 샘플마다 2개의 차원을 가진다. -> 보통 하나의 sample : 하나의 sequence -> cell의 입력은 샘플마다 타임스텝과 단어 표현으로 이루어진 2차원 배열이어야함** => sequence의 길이 : timestep의 길이

- 순환층은 기본적으로는 마지막 타임스탭의 은닉 상태만 출력으로 내보낸다.

> **ex) 20개의 timestep, 100개의 표현(특성)으로 이루어진 sample이 순환층의 cell을 통과하면 모든 타임스텝을 처리하고 난 후의 1차원의 은닉 상태만 출력된다. 이 은닉 상태의 크기는 셀에 있는 뉴런의 개수가 됨**

- 합성곱 신경망과 다른 점 : 마지막 셀의 출력이 1차원이기 때문에 Flatten 클래스로 펼칠 필요가 없음
